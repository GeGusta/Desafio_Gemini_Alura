```python
# Funções que serão utilizadas para realização das tarefas

# Função para ler o arquivo em pdf
def file_reader(file_name):
  try:
    pdf = pdfx.PDFx(file_name)
    return pdf.get_text()
  except Exception as e:
    print(f"Erro: Arquivo '{file_name}' fora do formato PDF")

# Função que faz o upload do arquivo
def file_upload():
  try:
    uploaded = files.upload()
    return list(uploaded.keys())[0]
  except Exception as e:
    print(f"Erro: Arquivo inválido")

# Função para formatar o texto em markdown
def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

# Função para realizar a tarefa de avaliar Currículo
def avaliar_curriculo():
  print ("Faça upload do seu currículo. (currículo.pdf) \n")
  curriculo = file_upload()

  if curriculo != None:
    file_content = file_reader(curriculo)
    if file_content != None:
      # Envio do currículo para o modelo
      print("\nRecebendo currículo...\n")
      resposta = recrutador.send_message("Tarefa 1 para o seguinte currículo: " + file_content)
      print("Analisando currículo\n")
      display(to_markdown(f'{resposta.text}'))

# Função que faz a tarefa de analisar o match do currículo com a descrição da vaga
def match_vaga():
  print ("Faça upload do seu currículo. (currículo.pdf) \n")
  curriculo = file_upload()
  vaga = input("Descrição da vaga: ")

  if curriculo != None and vaga != None:
    file_content = file_reader(curriculo)
    if file_content != None:
      # Envio do currículo e descrição da vaga para o modelo
      print("\nAnalisando o currículo e a vaga\n")
      resposta = recrutador.send_message("Tarefa 2 para do currículo: \n" + file_content + " e a seguinte descrição de vaga: \n" + vaga)
      display(to_markdown(f'{resposta.text}'))
```

# Fazendo a análise de Currículo

```python
# Inicialização da solução

print("Bem vindo ao avaliador de currículos e Match com a vaga pretendida. \n \nSelecione uma das opções a seguir: \n 1 - Avalie como está seu currículo \n 2 - Compare seu currículo com a descrição da vaga pretendida \n")
parar = 0

# Capta a opção escolhida
while parar == 0:
  opcao = input("Qual a opção escolhida: ")
  if opcao == "1" or opcao == "2":
    parar = 1
  else:
    parar = 0

# Chama a opção escolhida
if opcao == "1":
  avaliar_curriculo()

else:
  match_vaga()

print("\nObrigado!")
```
